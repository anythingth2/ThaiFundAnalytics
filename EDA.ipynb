# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Import Data

```python
fund_df = pd.read_csv('datasets/fund.csv')
fund_df
```

```python
nav_df = pd.read_csv('datasets/nav.csv', index_col='date', parse_dates=['date'])
nav_df
```

```python
unit_value_df = pd.read_csv('datasets/unit_value.csv', index_col='date', parse_dates=['date'])
unit_value_df
```

# Age of Fund

```python
fund_age_df = pd.DataFrame(nav_df.apply(lambda nav: nav.notna().sum())).rename(columns={0: 'fund_age_in_day'})
fund_df = fund_df.merge(fund_age_df, how='left', left_on='short_code', right_index=True)
fund_df
```

```python
px.histogram(fund_df, x='fund_age_in_day', nbins=32)
```

```python
fund_df.columns
```

```python
px.box(fund_df, x='aimc_category_name_th', y='fund_age_in_day', height=800)
```

## Filter only fully-developed fund

```python
IGNORED_SHORT_CODES = ['SCBFI3Y5']
fund_df = fund_df[~fund_df['short_code'].isin(IGNORED_SHORT_CODES)]
fund_df = fund_df[fund_df['fund_age_in_day'] > 365 * 2]
nav_df = nav_df[fund_df['short_code']]
unit_value_df = unit_value_df[fund_df['short_code']]
```

```python
(fund_df.groupby(['aimc_category_name_en'])
         .agg('count')['fund_id']
         .plot.bar(title='How many fund each group?',
                   figsize=(8, 6))
 
)
```

# Overview of fund in Thailand market

```python
by = ['aimc_broad_category_name_th', 'aimc_category_name_th']
df = fund_df.groupby(by).agg(count=('fund_id', 'count')).reset_index()
px.treemap(df,
            path=by,
            values='count', 
            color='count', 
            labels='count', 
            width=1200, 
            height=800,
            title='Number of fund per category'
        )
```

```python
by = ['dividend_policy','aimc_broad_category_name_th']
df = fund_df.groupby(by).agg(count=('fund_id', 'count')).reset_index()
px.treemap(df,
            path=by,
            values='count', 
            color='count', 
            labels='count', 
            width=1200, 
            height=800,
            title='Number of fund per category'
        )
```
